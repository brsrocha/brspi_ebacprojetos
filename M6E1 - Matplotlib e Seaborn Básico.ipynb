```python
# Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregar o arquivo
df = pd.read_csv(r'../SINASC_RO_2019.csv')
df
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# 2, i) Criando a figura 2x2
fig, axes = plt.subplots(1, 5, figsize=(20, 5))


# ii) Plotando em cada axes um gráfico com as variáveis necessárias
# ao invés de usar um gráfico de barras, usamos um countplot, devido a ausência de informações para o eixo y.'

sns.countplot(ax = axes[0], x='APGAR1', data=df)
sns.countplot(ax = axes[1], x='APGAR5', data=df)
sns.countplot(ax = axes[2], x='SEXO', data=df)
sns.countplot(ax = axes[3], x='GESTACAO', data=df)
sns.countplot(ax = axes[4], x='ESTCIVMAE', data=df)

# iii) Deixando o gráfico apresentável
plt.tight_layout()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# 3) Criando o gráfico de barras diretamente com o Seaborn usando a função
import numpy as np

# i) Criando a figure
plt.figure(figsize=(10, 6))

# ii) Usando o parâmetro estimator para calcular a média do APGAR5 por consulta
sns.barplot(data=df, x='CONSULTAS', y='APGAR5', estimator=np.mean, ci=None)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# 4, i) Criando a figure do gráfico
fig, axes = plt.subplots(2, 2, figsize=(10, 5))

# ii) criando cada countplot com as frequências das variáveis
sns.countplot(ax = axes[0,0], x='SEXO', data=df)
sns.countplot(ax = axes[0,1], x='GESTACAO', data=df)
sns.countplot(ax = axes[1,0], x='ESTCIVMAE', data=df)
sns.countplot(ax = axes[1,1], x='CONSULTAS', data=df)

# iii) Ajustando o layout
plt.tight_layout()
```
